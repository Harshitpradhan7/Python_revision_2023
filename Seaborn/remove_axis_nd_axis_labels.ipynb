```python
df = pd.DataFrame({'x': [1, 2, 3], 'y': [4, 5, 6]})

# Create subplots
fig, axes = plt.subplots(1, 3, figsize= (16,6))  # Adjust number of subplots as needed

for ax in axes.flat:
    df.plot(ax=ax)
    ax.set_xticklabels([])  # Remove x-axis tick labels
    ax.set_yticklabels([])  # Remove y-axis tick labels

plt.show()
```

```python
# Sample DataFrame
df = pd.DataFrame({'x': [1, 2, 3], 'y': [4, 5, 6]})

# Create subplots
fig, axes = plt.subplots(1, 3, figsize= (16,6))  # Adjust number of subplots as needed

for ax in axes.flat:
    df.plot(ax=ax)
    ax.get_xaxis().set_visible(False)  # Make x-axis invisible
    ax.get_yaxis().set_visible(False)  # Make y-axis invisible

plt.show()
```
